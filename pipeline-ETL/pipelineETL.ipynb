```python
import requests
import pandas as pd
```

```python
def extract():
    url = "https://jsonplaceholder.typicode.com/users"

    response = requests.get(url, timeout=10)
    response.raise_for_status()

    print ('Dados extraídos da API com sucesso')
    return response.json()

```

```python
def transform(data):
    # Normaliaza JSON
    df = pd.json_normalize(data)

    # Seleciona colunas relevantes
    df = df[[
        "id",
        "name",
        "username",
        "email",
        "address.city",
        "company.name"
    ]]

    # Renomeando as colunas 
    df.columns = [
        "id",
        "nome",
        "usuario",
        "email",
        "cidade",
        "empresa"
    ]

    # Padrnização
    df['nome'] = df['nome'].str.upper()
    df['cidade'] = df['cidade'].str.upper()

    # Coluna derivada
    df['dominio_email'] = df['email'].str.split("@").str[1]

    print('Dados transformados com sucesso')
    return df

   
   
```

```python
def load(df):
    df.to_csv('usuario_api_tratados.csv', index = False)
    print ('Dados carregados com sucesso')
```

```python
def pipeline_etl_api():
    dados_api = extract()
    df_tratado = transform(dados_api)
    load(df_tratado)

if __name__ == "__main__":
    pipeline_etl_api()
```
